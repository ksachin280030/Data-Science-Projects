```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
DATA_DIR = "C:/mini project/Classify clothes/images_original"
DATA_RAW = "C:/mini project/Classify clothes/images.csv"
WORK_DIR = "../"

```

```python
data = pd.read_csv(DATA_RAW)
data.head(10)
```

```python
data.info()
```

```python
data["sender_id"].nunique()
```

```python
plt.figure(figsize=(12, 4))
data["label"].value_counts(sort=True).plot(kind="bar")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data["kids"].value_counts(sort=True).plot(kind="pie", autopct="%0.2f")
plt.show()
```

```python
fig = plt.figure(figsize=(16, 12))
for i in range(12):
    fig.add_subplot(3, 4, i+1)
    image = plt.imread(os.path.join(DATA_DIR, data["image"][i]+".jpg"))
    plt.imshow(image)
plt.show()
```

```python
train , test = train_test_split(data, stratify=data["label"], test_size=0.2, shuffle=True, random_state=42)
train.to_csv(os.path.join(WORK_DIR, "train.csv"), index=False)
test.to_csv(os.path.join(WORK_DIR, "test.csv"), index=False)
```
