```python
for col in data.columns:
    if data[col].dtype != 'O':
        data[col].fillna(data[col].mean(), inplace = True)
    else:
        data[col].fillna(data[col].mode()[0], inplace = True)
```

```python
data.info
```

```python
plt.figure(figsize = (8,5))
ax = sns.countplot(data.Species, hue = data.Island)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Number of Species", fontsize = 20);
```

```python
plt.figure(figsize = (8,5))
sns.kdeplot(data["Body Mass (g)"], hue = data.Species, shade = True)
plt.title("Body Mass Histogram", fontsize = 20);
```
